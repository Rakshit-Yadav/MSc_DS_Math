```python
def LogisticMap(x,r):
    return r*x*(1.0-x)
# the above function is the formula for the logistic map. this will produce
# different values of x_i.

```

```python
nSteps = 100
r = 3.9
t = range(0,nSteps+1)
x = np.zeros(n+1) # this is to create an array using numpy library and this array will get updated afterwards
x[0] = 0.01 # choosing the initial starting value of x from 0 to 1
for i in range(1, n+1):
    x[i] = LogisticMap(x[i-1],r) # this generates the different values of x between 0 to 1

```

```python
#Plotting the logistic map
plt.plot(t, x, '-', linewidth=1,color='black') # for actually plotting
# '-' means solid line, t and x are paramters and other are self-explanatory

plt.scatter( t, x, color="darkviolet" ) # add dots at the points

plt.title('Logistic Map: r=' + str(r))

plt.xlabel('Step, n')

plt.ylabel('x(n)')

plt.show()
```
