## some definitions to learn
### OBJECT: 
#####         In Python, everything is an object.
#####         variable, function, class, pretty much anything else.
### CLASS: 
#####        A Structure for defining a type of object.
#####        dict, list, all are classes. used for grouping data together,
#####        and haing variables behave a particular, repeatable way.
### METHOD: 
#####         A function that is associated/part of a class.
#####         e.g.- .append() for a list is a method of the class List.
### INSTANCE: 
#####           An object of a particular class. e.g.- i made a list "nice".
#####           nice is an instance of the class List.
### ATTRIBUTE: 
#####           class attributes and instance attributes. assigning and 
#####           naming different variables, and binding it to an instance
#####            or the class itself.

## An Example class from lab notes

```python
class Dog:
    # Define some class variables
    # These variables apply to all instances
    # If we have a dog with a different number of legs, 
        # this could be an instance attribute instead
    num_legs = 4
    
    # This is basically the constructor
    # When we create an instance, this __init__ function is run
    # Like any function, it takes arguments
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
        # We have some attributes that we might not know at the start
        # But we write them here so we know all attributes an instance has
        # So we set them to `None` for now
        self.gender = None
        self.colour = None
        self.breed = None
    
    def bark(self):
        print(self.name, "woofs! That's their way of saying hello :)")
    
    # We can have different actions bases on the attributes of the instance
    def fetch(self):
        if self.age > 10:
            print(self.name, "is too old to play fetch!")
        else:
            print(self.name, "runs after the stick like a bullet!")
            
    def check_gender(self, gender):
        print(self.name, "is", gender, "!")
        # Assign the given gender argument as a attribute for this instance
        self.gender = gender
    
# Create an instance
a_dog = Dog("Fido", 11)

# Let's look at class and attribute variables
print(Dog.num_legs) # The class variable we set
print(a_dog.num_legs) # Instances inherit all class variables
a_dog.num_legs = 3
print(Dog.num_legs) # the class variable is unchanged
print(a_dog.num_legs) # this instance variable is now different

print(a_dog.bark()) # Prints our statement and returns None by default
a_dog.check_gender("male") # Sets the gender of the dog
print(a_dog.gender) # prints male
a_dog.fetch() # Poor Fido is too old to play fetch :(
```

```python
# defining class
class Point:
    
    # the constructor with attributes as x axis and y axis point given by the user
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    # this function only moves the point
    def move_point(self, delta_x, delta_y):
        self.x += delta_x
        self.y += delta_y
    
    #this function moves the point and also tells us the new point
    def move_point_project(self,delta_x,delta_y):
        x = self.x + delta_x
        y = self.y + delta_y
        Point.check_point(x,y)
    
    # this staticmethod function is for checking if the moved projection of the
    # point is within range [0,10](inclusive) or not.
    @staticmethod
    def check_point(x,y):
        if (x<0 or x>10) or (y<0 or y>10): # the brackets on x and y evaluates them
                                        # first and then evaluates the rest
            print("Point moved out of the range [0,10]")
            
        else:
            # this gives us the printing of the new projected point
            print("moving constraints are valid")
            print("The new point is: ",x,"and",y)
            

a_point = Point(7,4)
print("Starting point:", a_point.x, a_point.y)
a_point.move_point(-1, 5)
print("After moving -1 and +5:", a_point.x, a_point.y)
a_point.move_point_project(2,3)
print("\n")
a_point.move_point_project(1,0)
```
