# Exercise - 1

```python
# we ask the user to input any 2 decimal numbers/floats
var1 = float(input("Please give a decimal number: "))
var2 = float(input("Please give another decimal number: "))

# then we use if-else to tell the user which input is bigger
if var1 > var2:
    print("Your first input",var1,"is greater than your second input",var2)
else:
    print("Your second input",var2,"is greater than your first input",var1)

# we can also use min_value and max_value to store the minimum and
# maximum values and then display it using if-else which will be a 
# better option if we want to use this information afterwards.
```

# Exercise - 2

```python
# fist prompting the user to input 3 interests
interest_1 = input("What are you interested in: ")
interest_2 = input("What are you interested in: ")
interest_3 = input("What are you interested in: ")

# now printing these 3 interests as said in the exercise
print("You are interested in",interest_1,',',interest_2,'and',
      interest_3)

# Now we modify our code so that the user can input however many 
# interests that they want to add

interests = [] # making an empty list
number_of_interests = int(input("How many interests would you like to add: "))

for i in range(number_of_interests): #using for loop and running it for
                                    # a range of number of interests
    interests.append(input("What are you interested in: "))
print(interests) # this is the resultant list
```
