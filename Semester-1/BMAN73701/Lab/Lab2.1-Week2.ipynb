### Exceptions and Error Handling

```python
# try contains code 
try:
    x = float(input("Please provide a numnber: "))
    inverse = 1/x
    print(inverse)
    
# we use except to either print something when we face an error, i.e., an exception like in this situation
except ValueError:
    print("Incorrect type")
except ZeroDivisionError:
    print("Can't divide by 0")
finally:
    # This code is run no matter what
    print("Something happened...who knows what!") # this will run no matter what, 'finally' is used for that
```

#### isinstance function to check variable type

```python
a = 4
b = 5.3
c = "a string"
d = None

def check_var(var):
    if isinstance(var, (int,float)):
        print("It's a number!")
    elif isinstance(var, str):
        print("It's a string!")
    else:
        print("I'm not sure what it is...")
        
check_var(a)
check_var(b)
check_var(c)
check_var(d)

#we can use this to also prompt the user if they have input the wrong type of data and to ask for input
#again or in cases of int or float, just convert them
```

# Exercise - 1

```python
# this is the recursive approach
def iterative_factorial(n):
    if n==0:
        return 1
    else:
        return n*iterative_factorial(n-1)
print(iterative_factorial(5))
```

```python
# this is the non-recursive approach using loops
def iterative_factorial(n):
    try:
        a =1 
        for i in range(2,n+1,):
            a *= i
        return a
    except TypeError:
        print("Your input is not an integer")
            
print(iterative_factorial(5))
print(iterative_factorial("ni"))
```

# Exercise - 2

```python
def odd_list_product(nice_list):
    if nice_list == []:
        return 0
    elif len(nice_list) <= 2:
        return nice_list[0]
    else:
        return nice_list[0] * odd_list_product(nice_list[2:])
print(odd_list_product([1,2,3,4,5,6,7,8]))
```

# Exercise - 3

```python
def vowel_checker(nice):
    result = []
    for i in nice:
        if i in ['a','e','i','o','u','A','E','I','O','U']:
            result.append(i)
        else:
            continue
    return result
vowel_checker("Tom the rAt")
```
