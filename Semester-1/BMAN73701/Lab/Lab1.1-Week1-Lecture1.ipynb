## c) Calculate the positive root of the following equation:
## 34*x^2+68*xâˆ’510

```python
# first we assign the values by considering it with the standard
# equation a * x^2 + b * x + c
a = 34
b = 68
c = -510

# now the basic solution for a quadratic equation is:
# x = (-b + square root(4*a*c))/(2*a)

# to use the square root there are 2 methods
# 1. use **0.5
# 2. import math module and use .sqrt function

import math
x = (-b+math.sqrt(b**2 - (4*a*c)))/(2*a)
print("The positive root of the equation is:",x)
```
